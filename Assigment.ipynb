```python
class Vehicle:
    # Constructor to initialize attributes
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    # Method to display general vehicle information
    def display_info(self):
        return f"Vehicle: {self.year} {self.make} {self.model}"
```

```python
class Car(Vehicle):
    # Constructor for Car, calling the Parent's constructor
    def __init__(self, make, model, year, doors):
        # Call the parent class's constructor
        super().__init__(make, model, year)
        # Add a new attribute specific to the Car class
        self.doors = doors

    # Method specific to the Car class
    def honk(self):
        return "Beep! Beep!"

    # Override the parent's method (Optional)
    def display_info(self):
        # Use the parent's display_info and add more details
        parent_info = super().display_info()
        return f"{parent_info}, Doors: {self.doors}"
```
